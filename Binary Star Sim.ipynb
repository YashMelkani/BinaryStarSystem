```python
#########################################################THIS WORKS##############################################################
class binarySystem3:
    def __init__(self, inc, mass1, mass2, rad1, rad2, temperature1, temperature2, samples): # samples is the number of data points and r1 is the larger star
        self.m1 = mass1
        self.m2 = mass2 
        self.i = m.radians(inc)
        self.r1 = rad1 
        self.r2 = rad2 
        self.temp1 = temperature1 
        self.temp2 = temperature2
        self.s = samples
        
    def graph(self):
        q = self.m2/self.m1 
        distance = 1 # distance between stars in 3D space
        
        #Fluxes
        f1 = 5.67037321212121e-8 * (self.temp1**4)
        f2 = 5.67037321212121e-8 * (self.temp2**4)
        
        duration = 2 * m.pi
        interval = duration / self.s 
        phase = [] # X-axis
        brightness = [] # Y-axis
        intensity = 0 #temporary assignment
        maxB = (m.pi*f1*self.r1**2) + (m.pi*f2*self.r2**2) #Maximum brightness
        theta = 3*m.pi/2 # phase angle
        while theta <= duration + 3*m.pi/2:
            X1 = distance * m.sin(theta)
            Y1 = distance * m.cos(self.i) * m.cos(theta)
            Z1 = distance * m.sin(self.i) * m.cos(theta) 
            
            X2 = distance * m.sin(theta)
            Y2 = distance * m.cos(self.i) * m.cos(theta)
            Z2 = distance * m.sin(self.i) * m.cos(theta)
            
            x1 = -X1 / (1 + 1/q)
            y1 = -Y1 / (1 + 1/q)
            z1 = -Z1 / (1 + 1/q)
            
            x2 = X2/(1 + q)
            y2 = Y2/(1 + q)
            z2 = Z2/(1 + q)
            
            distXY = (((x1-x2)**2) + ((y1-y2)**2)) ** 0.5 #apparent distance between the stars
            
            if distXY >= (self.r1 + self.r2): # if they're not eclipsing
                a1 = m.pi * self.r1**2
                a2 = m.pi * self.r2**2
                intensity = a1*f1 + a2*f2
                
            
            elif (self.r1 + self.r2) > distXY and distXY >= abs(self.r1-self.r2) and z1 > z2: # if r1 is eclipsing r2 but not completely
                
                if distXY >= (self.r1**2 - self.r2**2) ** 0.5: # Shallow Eclipse
                    covAngle1 = 2 * m.acos((distXY**2 + self.r1**2 - self.r2**2)/(2*distXY*self.r1))
                    covAngle2 = 2 * m.acos((distXY**2 + self.r2**2 - self.r1**2)/(2*distXY*self.r2))
                    deltaA1 = (0.5 * self.r1**2) * (covAngle1 - m.sin(covAngle1))
                    deltaA2 = (0.5 * self.r2**2) * (covAngle2 - m.sin(covAngle2))
                    a1 = m.pi * self.r1**2 
                    a2 = m.pi * self.r2**2 - deltaA1 - deltaA2
                    intensity = a1*f1 + a2*f2

                elif distXY < (self.r1**2 - self.r2**2) ** 0.5: # Deep Eclipse
                    covAngle1 = 2 * m.acos((distXY**2 + self.r1**2 - self.r2**2)/(2*distXY*self.r1))
                    covAngle2 = 2 * m.pi - (2 * m.acos((distXY**2 + self.r2**2 - self.r1**2)/(2*distXY*self.r2)))
                    deltaA1 = (0.5 * self.r1**2) * (covAngle1 - m.sin(covAngle1))
                    deltaA2 = (0.5 * self.r2**2) * (covAngle2 - m.sin(covAngle2))
                    a1 = m.pi * self.r1**2 
                    a2 = deltaA2 - deltaA1
                    intensity = a1*f1 + a2*f2               
                
                
            elif (self.r1 + self.r2) > distXY and distXY > abs(self.r1-self.r2) and z2 > z1: # if r2 is eclipsing r1 but not completely
                
                if distXY >= (self.r1**2 - self.r2**2) ** 0.5: # Shallow Eclipse
                    covAngle1 = 2 * m.acos((distXY**2 + self.r1**2 - self.r2**2)/(2*distXY*self.r1))
                    covAngle2 = 2 * m.acos((distXY**2 + self.r2**2 - self.r1**2)/(2*distXY*self.r2))
                    deltaA1 = (0.5 * self.r1**2) * (covAngle1 - m.sin(covAngle1))
                    deltaA2 = (0.5 * self.r2**2) * (covAngle2 - m.sin(covAngle2))
                    a1 = (m.pi * self.r1**2) - deltaA1 - deltaA2
                    a2 = m.pi * self.r2**2
                    intensity = a1*f1 + a2*f2
                    
                elif distXY < (self.r1**2 - self.r2**2) ** 0.5: # Deep Eclipse
                    covAngle1 = 2 * m.acos((distXY**2 + self.r1**2 - self.r2**2)/(2*distXY*self.r1))
                    covAngle2 = 2 * m.pi - (2 * m.acos((distXY**2 + self.r2**2 - self.r1**2)/(2*distXY*self.r2)))
                    deltaA1 = (0.5 * self.r1**2) * (covAngle1 - m.sin(covAngle1))
                    deltaA2 = (0.5 * self.r2**2) * (covAngle2 - m.sin(covAngle2))
                    a1 = (m.pi * self.r1**2) - (m.pi * self.r2**2) + deltaA2 - deltaA1  
                    a2 = m.pi * self.r2**2
                    intensity = a1*f1 + a2*f2
                
                
            elif distXY < abs(self.r1 - self.r2) and z1 > z2: #if r1 is completely eclipsing r2
                a1 = m.pi * self.r1**2
                a2 = 0
                intensity = a1*f1 + a2*f2  
        
                
            elif distXY < abs(self.r1 - self.r2) and z2 > z1: #if r2 is completely eclipsing r1
                a1 = (m.pi * self.r1**2) - (m.pi * self.r2**2)
                a2 = m.pi * self.r2**2
                intensity = a1*f1 + a2*f2
        
                
                
            elif z2 == 0 and z1 == 0: # inclincation angle is zero --> z1 = z2 = 0
                a1 = m.pi * self.r1**2
                a2 = m.pi * self.r2**2
                intensity = a1*f1 + a2*f2
                
            phase.append((theta)/(2 * m.pi) - 0.75)
            brightness.append(intensity/maxB)
            theta += interval
            # print(distXY)
        
        plt.scatter(phase,brightness, s=10)
        plt.ylim([0,1])
        plt.show()
        
        
sys = binarySystem3(82.3, 1, 0.42, 0.349, 0.282, 9750, 5600, 200)
sys.graph()
```
